```python

!pip install opencv-python-headless


import numpy as np
import matplotlib.pyplot as plt
import cv2

#  Step 1: Generate a grayscale gradient image (256x256)
img = np.tile(np.arange(0, 256, dtype=np.uint8), (256, 1))  # Horizontal gradient
img_color = cv2.merge([img, img, img])  # Convert to 3-channel grayscale (RGB)

#  Step 2: Define encryption and decryption using XOR
def encrypt_image(image, key=101):
    return np.bitwise_xor(image, key)

def decrypt_image(encrypted_image, key=101):
    return np.bitwise_xor(encrypted_image, key)

# Step 3: Apply encryption and decryption
encrypted_img = encrypt_image(img_color)
decrypted_img = decrypt_image(encrypted_img)

#  Step 4: Plot images + histograms
fig, axs = plt.subplots(3, 2, figsize=(12, 7))

#  Original
axs[0, 0].imshow(cv2.cvtColor(img_color, cv2.COLOR_BGR2RGB))
axs[0, 0].set_title("Original Image")
axs[0, 0].axis('off')
axs[0, 1].hist(img_color.ravel(), bins=256, color='gray')
axs[0, 1].set_title("Original Histogram")

#  Encrypted
axs[1, 0].imshow(cv2.cvtColor(encrypted_img, cv2.COLOR_BGR2RGB))
axs[1, 0].set_title("Encrypted Image (XOR)")
axs[1, 0].axis('off')
axs[1, 1].hist(encrypted_img.ravel(), bins=256, color='orange')
axs[1, 1].set_title("Encrypted Histogram")

#  Decrypted
axs[2, 0].imshow(cv2.cvtColor(decrypted_img, cv2.COLOR_BGR2RGB))
axs[2, 0].set_title("Decrypted Image")
axs[2, 0].axis('off')
axs[2, 1].hist(decrypted_img.ravel(), bins=256, color='green')
axs[2, 1].set_title("Decrypted Histogram")

plt.tight_layout()
plt.show()
```

```python
#  Install OpenCV if needed
# !pip install opencv-python-headless

import numpy as np
import cv2
import matplotlib.pyplot as plt

#  Step 1: User Input
print("Select image type:")
print("1 - Colored Blocks")
print("2 - Grayscale Gradient")
print("3 - Random Noise")
img_choice = int(input("Enter choice (1/2/3): "))

key = int(input("Enter encryption key (0255): "))
print("\nSelect encryption method:")
print("1 - XOR")
print("2 - Invert Colors (255 - pixel)")
print("3 - Swap Red and Blue Channels")
enc_choice = int(input("Enter encryption method (1/2/3): "))

#  Step 2: Generate selected image
if img_choice == 1:
    img = np.zeros((100, 100, 3), dtype=np.uint8)
    img[:50, :50] = [255, 0, 0]     # Red
    img[:50, 50:] = [0, 255, 0]     # Green
    img[50:, :50] = [0, 0, 255]     # Blue
    img[50:, 50:] = [255, 255, 0]   # Yellow
elif img_choice == 2:
    gray = np.tile(np.arange(0, 256, dtype=np.uint8), (256, 1))
    img = cv2.merge([gray, gray, gray])
    img = cv2.resize(img, (100, 100))
else:
    img = np.random.randint(0, 256, (100, 100, 3), dtype=np.uint8)

#  Step 3: Encryption methods
def encrypt_xor(image, key):
    return np.bitwise_xor(image, key)

def invert_colors(image):
    return 255 - image

def swap_rb(image):
    return image[..., ::-1]  # Swap R and B

#  Step 4: Apply Encryption
if enc_choice == 1:
    encrypted_img = encrypt_xor(img, key)
    decrypted_img = encrypt_xor(encrypted_img, key)
elif enc_choice == 2:
    encrypted_img = invert_colors(img)
    decrypted_img = img
else:
    encrypted_img = swap_rb(img)
    decrypted_img = img

#  Step 5: Visualizations
fig, axs = plt.subplots(3, 2, figsize=(14, 10))

def plot_image_and_hist(ax_img, ax_hist, image, title):
    ax_img.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    ax_img.set_title(title)
    ax_img.axis('off')
    for i, color in enumerate(['b', 'g', 'r']):
        ax_hist.hist(image[:, :, i].ravel(), bins=256, color=color, alpha=0.5)
    ax_hist.set_title(f"{title} - RGB Histogram")

plot_image_and_hist(axs[0][0], axs[0][1], img, "Original Image")
plot_image_and_hist(axs[1][0], axs[1][1], encrypted_img, "Encrypted Image")
plot_image_and_hist(axs[2][0], axs[2][1], decrypted_img, "Decrypted Image")

plt.tight_layout()
plt.show()
```
